```python
import pandas as pd
import datetime
from random import randint
```

```python
chats = pd.read_csv('input/rdany_conversations_2016-03-01.csv')
```

```python
chats = chats[~chats['text'].str.startswith('[')]
```

```python
chats.sort_values(by=['hashed_chat_id','date']).head()
```

```python
chat_id = list(set(list(chats['hashed_chat_id'])))
chat_id = {chat_id[i]:i for i in range(len(chat_id))}
chats['hashed_chat_id'] = chats['hashed_chat_id'].apply(lambda x: chat_id.get(x,x))

message_id = list(set(list(chats['hashed_message_id'])))
message_id = {message_id[i]:i for i in range(len(message_id))}
chats['hashed_message_id'] = chats['hashed_message_id'].apply(lambda x: message_id.get(x,x))
```

```python
base = datetime.datetime.now()
date_list = [base - datetime.timedelta(minutes=(10*x-randint(0,9)),seconds=randint(1,59),milliseconds=randint(1,99999)) for x in range(len(chats['date']))][::-1]
date_list
```

```python
chats = chats.sort_values(by='date')
chats['date'] = date_list
```

```python
chats.sort_values(by='hashed_chat_id')
```
